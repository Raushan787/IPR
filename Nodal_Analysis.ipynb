```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IPR dataset.png

```python
#Uploading the IPR dataset (From PetroWiki)
q_ipr=[0,1999,3094,3902,4512,4963,5275,5458,5519]
pwf_ipr=[4000,3500,3000,2500,2000,1500,1000,500,14.65]

 #Creating dataframe using pandas
df_ipr=pd.DataFrame({"Pwf":pwf_ipr,
                      "Q":q_ipr})
df_ipr
```

```python
#Plotting Inflow Performance Relation (IPR) Curve

plt.figure(figsize=(10,8)) #8->horizontal and 6 vertical length
plt.plot(df_ipr["Q"],df_ipr["Pwf"],lw=5)
plt.title("IPR Curve")
plt.xlabel(" Q (Mscf/d)")
plt.ylabel(" Pwf (psia)")
plt.grid(True)
plt.show()
```

TPR dataset.png

```python
#Uploading TPR dataset(PetroWiki)
q_tpr=np.arange(1000,6500,500)
p_190=[1334,1400,1487,1592,1712,1843,1984,2132,2287,2446,2609]
p_2375=[1298,1320,1351,1390,1435,1487,1545,1609,1677,1749,1824]
p_2875=[1286,1294,1305,1319,1336,1356,1378,1403,1431,1461,1493]

#Creating Dataframe
tpr=pd.DataFrame({"Q":q_tpr,
                  "Pwf(1.9 tbg)":p_190,
                  "Pwf(2.375 tbg)":p_2375,
                  "Pwf(2.875 tbg)":p_2875})
tpr
```

```python
#Plotting Tubing Performance Relation (TPR) Curve
plt.figure(figsize=(12,6))
plt.plot(tpr["Q"],tpr["Pwf(1.9 tbg)"], label="TPR for 1.9 tubing ",linewidth=3)
plt.plot(tpr["Q"],tpr["Pwf(2.375 tbg)"], label="TPR for 2.375 tubing ",linewidth=3)
plt.plot(tpr["Q"],tpr["Pwf(2.875 tbg)"], label="TPR for 2.875 tubing ",linewidth=3)
plt.xlabel(" Q (Mscf/d)")
plt.ylabel(" Pwf (psia)")
plt.xlim(0,6500)
plt.ylim(0,3000)
plt.title("Tubing Performance Relation (TPR) Curve")
plt.grid(True)
plt.legend(loc="best")
```

**Equation:**

P_loss (flow from reservoir to bottom hole) = Pr - Pwf  --> IPR

P_loss (flow from bottom hole to surface)   = Pwf - Pwh --->TPR

**According to nodal analysis **

Pwf(IPR)=Pwf(TPR)

**P_loss (flow from bottom hole to surface) +Pwh =Pr - P_loss (flow from reservoir to bottom hole) **

```python
#Plot of IPR and TPR
plt.figure(figsize=(12,8))
plt.style.use("classic")
plt.plot(df_ipr["Q"],df_ipr["Pwf"],label="IPR Curve",lw=5)
plt.plot(tpr["Q"],tpr["Pwf(1.9 tbg)"], label="TPR for 1.9 tubing ",linewidth=3)
plt.plot(tpr["Q"],tpr["Pwf(2.375 tbg)"], label="TPR for 2.375 tubing ",linewidth=3)
plt.plot(tpr["Q"],tpr["Pwf(2.875 tbg)"], label="TPR for 2.875 tubing ",linewidth=3)
plt.xlim(0,6500)
plt.ylim(0,4500)
plt.xlabel(" Q (Mscf/d)")
plt.ylabel(" Pwf (psia)")
plt.title("Plot of IPR and TPR Curve ")
plt.grid(True)
plt.legend()
```
