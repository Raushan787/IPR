```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IPR dataset.png

```python
#Uploading the IPR dataset (From PetroWiki)
q_ipr=[0,1999,3094,3902,4512,4963,5275,5458,5519]
pwf_ipr=[4000,3500,3000,2500,2000,1500,1000,500,14.65]

df_ipr=pd.DataFrame({"Pwf":pwf_ipr,
                      "Q":q_ipr})
df_ipr
```

```python
#Plotting Inflow Performance Relation (IPR) Curve
plt.figure(figsize=(8,4))
x_ipr=df_ipr["Q"].values
y_ipr=df_ipr["Pwf"].values

plt.figure(figsize=(8,4)) #8->horizontal and 4 vertical length
plt.plot(x_ipr,y_ipr)
plt.title("IPR Curve")
plt.xlabel(" Q (Mscf/d)")
plt.ylabel(" Pwf (psia)")
plt.grid(True)
plt.show()
```

TPR dataset.png

```python
#Uploading TPR dataset(PetroKIwi)
q_tpr=np.arange(1000,6500,500)
p_190=[1334,1400,1487,1592,1712,1843,1984,2132,2287,2446,2609]
p_2375=[1298,1320,1351,1390,1435,1487,1545,1609,1677,1749,1824]
p_2875=[1286,1294,1305,1319,1336,1356,1378,1403,1431,1461,1493]

tpr=pd.DataFrame({"Q":q_tpr,
                  "Pwf(1.9 tpg)":p_190,
                  "Pwf(2.375 tpg)":p_2375,
                  "Pwf(2.875 tbg)":p_2875})
tpr
```

```python
#Plotting Tubing Performance Relation (TPR) Curve
plt.figure(figsize=(8,4))
plt.plot(q_tpr,p_190, label="TPR for 1.9 tubing ",linewidth=2)
plt.plot(q_tpr,p_2375, label="TPR for 2.375 tubing ",linewidth=2)
plt.plot(q_tpr,p_2875, label="TPR for 2.875 tubing ",linewidth=2)
plt.xlabel(" Q (Mscf/d)")
plt.ylabel(" Pwf (psia)")
plt.title("Tubing Performance Relation (TPR) Curve")
plt.grid(True)
plt.legend()
```

**Equation:**

Pwf(IPR)=Pe - delP_r

Pwf(TPR)=P_sep +  delP_t + delP_sl

**According to nodal analysis **

Pwf(IPR)=Pwf(TPR)

**Pe - delP_r=P_sep +  delP_t + delP_sl**

```python
#Plotting IPR curve and TPR curve on same graph
plt.figure(figsize=(8,4))
plt.style.use("fivethirtyeight")

plt.plot(x_ipr,y_ipr,label="IPR",linewidth=2)

plt.plot(q_tpr,p_190, label="TPR for 1.9 tubing ",linewidth=2)
plt.plot(q_tpr,p_2375, label="TPR for 2.375 tubing ",linewidth=2)
plt.plot(q_tpr,p_2875, label="TPR for 2.875 tubing ",linewidth=2)

plt.xlabel("Flow rate  (Mscf/d)")
plt.ylabel("Flowing BHP (Psia)")
plt.title("Nodal Analysis for a gas well")

plt.legend(loc="best")
```
